## 1. What are the highest-paying jobs?

```python
import pandas as pd

data = {
    'job_title': ['Data Scientist', 'Software Engineer', 'ML Engineer', 
                  'Data Engineer', 'Data Scientist', 'ML Engineer', 
                  'Software Engineer', 'Data Analyst'],
    'salary_in_usd': [120000, 95000, 130000, 110000, 125000, 135000, 100000, 85000]
}

jobs_data = pd.DataFrame(data)

highest_paying_jobs = (
    jobs_data.groupby('job_title')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Highest-Paying Jobs:\n", highest_paying_jobs)
```

## 3. How does salary change for entry-level, mid-level, and senior jobs?

## 9. Do bigger companies pay more than smaller ones?

```python
import pandas as pd

data = {
    'company_size': ['S', 'M', 'L', 'L', 'M', 'S', 'L', 'M', 'S', 'L'],
    'salary_in_usd': [50000, 70000, 120000, 110000, 75000, 
                      55000, 130000, 77000, 52000, 125000]
}

jobs_data = pd.DataFrame(data)

company_size_map = {'S': 'Small', 'M': 'Medium', 'L': 'Large'}
jobs_data['company_size_mapped'] = jobs_data['company_size'].map(company_size_map)

salary_by_company_size = jobs_data.groupby('company_size_mapped')['salary_in_usd'].mean()

print("Average Salary by Company Size:\n", salary_by_company_size)
```

## 1. Average Salaries by Job Title (Bar Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'job_title': ['Data Scientist', 'Software Engineer', 'ML Engineer', 
                  'Data Analyst', 'Product Manager', 'Backend Engineer', 
                  'Data Engineer', 'DevOps Engineer', 'Team Lead', 'Data Scientist'],
    'salary_in_usd': [120000, 95000, 130000, 85000, 115000, 98000, 
                      110000, 105000, 125000, 118000]
}

jobs_data = pd.DataFrame(data)

avg_salaries_by_title = jobs_data.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_salaries_by_title.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Salaries by Job Title', fontsize=14)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjusts layout to prevent label cutoff
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 2. Salary Trends by Experience Level (Line Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'experience_level': ['Entry', 'Mid', 'Senior', 'Entry', 'Mid', 
                         'Senior', 'Entry', 'Senior', 'Mid', 'Entry'],
    'salary_in_usd': [50000, 70000, 120000, 55000, 75000, 
                      130000, 52000, 125000, 78000, 53000]
}

jobs_data = pd.DataFrame(data)

salary_by_experience = jobs_data.groupby('experience_level')['salary_in_usd'].mean()

plt.figure(figsize=(8, 5))
plt.plot(salary_by_experience.index, salary_by_experience.values, marker='o', linestyle='-', color='blue')
plt.title('Salary Trends by Experience Level', fontsize=14)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.grid()
plt.show()
```

## 3. Distribution of Salaries (Histogram)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'salary_in_usd': [120000, 95000, 130000, 110000, 125000, 
                      135000, 100000, 85000, 140000, 105000, 
                      95000, 90000, 115000, 130000, 125000]
}

jobs_data = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.hist(jobs_data['salary_in_usd'], bins=20, edgecolor='black', color='lightblue')
plt.title('Distribution of Salaries', fontsize=14)
plt.xlabel('Salary (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 4. Employment Types (Pie Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'employment_type': ['Full-Time', 'Contract', 'Part-Time', 'Full-Time', 'Freelance', 
                        'Contract', 'Full-Time', 'Freelance', 'Part-Time', 'Full-Time']
}

jobs_data = pd.DataFrame(data)

employment_counts = jobs_data['employment_type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    employment_counts, 
    labels=employment_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['skyblue', 'lightgreen', 'orange', 'pink']
)
plt.title('Proportion of Employment Types', fontsize=14)
plt.show()
```

## 5. Company Size vs. Salary (Scatter Plot)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'company_size': ['S', 'M', 'L', 'L', 'M', 'S', 'L', 'M', 'S', 'L'],
    'salary_in_usd': [50000, 70000, 120000, 110000, 75000, 
                      55000, 130000, 77000, 52000, 125000]
}

jobs_data = pd.DataFrame(data)

company_size_map = {'S': 1, 'M': 2, 'L': 3}
jobs_data['company_size_numeric'] = jobs_data['company_size'].map(company_size_map)

plt.figure(figsize=(10, 6))
plt.scatter(jobs_data['company_size_numeric'], jobs_data['salary_in_usd'], alpha=0.5, color='purple')
plt.title('Company Size vs Salary', fontsize=14)
plt.xlabel('Company Size (1=Small, 2=Medium, 3=Large)', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.grid()
plt.show()
```

## 6. Geographic Salary Heatmap (Optional Visualization)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'company_location': ['US', 'UK', 'US', 'CA', 'UK', 
                         'DE', 'CA', 'US', 'DE', 'CA'],
    'salary_in_usd': [120000, 90000, 150000, 100000, 95000, 
                      110000, 105000, 130000, 115000, 108000]
}

jobs_data = pd.DataFrame(data)

salaries_by_country = jobs_data.groupby('company_location')['salary_in_usd'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=salaries_by_country, x='company_location', y='salary_in_usd', palette='coolwarm')
plt.title('Average Salaries by Country', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
